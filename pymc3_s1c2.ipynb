```python
import numpy as np
import pymc3 as pm
import arviz as az
import matplotlib.pyplot as plt
import corner
```

```python
### Constants
G = 4.301e-9
# G = 4.49975e-6
H = 67.6
Dc = 200.
b, g = 2., 3. 
```

```python
### Functions
def c2_dz(Mh):
    return 11.5 * (1. + np.power(Mh/(1.05e12), 0.65))

def Vvir(Mh):
    return np.sqrt((Dc*(H)**2/2.)**(1./3.) * (G*Mh)**(2./3.))

def Mvir(Vh):
    return np.sqrt(2./Dc)/G/H * Vh**3

def Rvir(Mh):
    rho_c = 3. * (H)**2 / (8. * np.pi * G)
    rho_hat = 4. / 3. * np.pi * Dc * rho_c
    return 1e3 * np.power(Mh / rho_hat, 1./3.)

def params_from_s1c2(s1,c2,r1,rvir,mvir):
    x12=np.sqrt(r1/rvir)
    c12=np.sqrt(c2)
    a=(1.5*s1-2.*(3.5-s1)*x12*c12)/(1.5-(3.5-s1)*x12*c12)
    c=((s1-2.)/((3.5-s1)*x12-1.5/c12))**2.
    return (c, a, b, g, rvir, mvir)

def mu(c, a, b, g):
    return np.power(1.+np.power(c,1./b),b*(g-a))/np.power(c,3.-a)

# circular velocity of Dekel Zaho halo
def vc_dz(r, Mh, s1, c2):
#     G_kpc_gyr = G * 0.957e-3
    rvir = Rvir(Mh)
    r1 = 0.01 * rvir
    cc, a, b, g, rvir, Mh = params_from_s1c2(s1,c2,r1,rvir,Mh)
    rs = rvir/cc
    x = r/rs
#     Vvir2 = G_kpc_gyr*Mh/rvir
    Vvir2 = Vvir(Mh)**2
    return np.sqrt(Vvir2*mu(cc, a, b, g)*cc*np.power(x,2.-a)/np.power(1.+np.power(x,1./b),b*(g-a)))

### c function (NFW)
def c_nfw(M): # from Dutton & Maccio' (2014) -- with c_200
    return 10.**(0.905 - 0.101 * (np.log10(M*H/100.)-12))

def vc_nfw(R, Mh, cc):
    return np.sqrt(Vvir(Mh)**2*Rvir(Mh)/R*fc(cc*R/Rvir(Mh))/fc(cc))

def fc(x):
    return np.log(1+x)-x/(1+x)
```

```python
rotcurve = np.loadtxt("NGC5055_rotmod.dat", unpack=True, comments='#')

with pm.Model() as dz_model:
    data  = pm.Data('data', rotcurve)
    logMh = pm.Uniform('logMh', lower=10., upper=15., testval=12.)
    logs1 = pm.Uniform('logs1', lower=-2., upper=2., testval=1.)
    logml = pm.Uniform('logml', lower=-2., upper=2., testval=-0.3)
    logc2 = pm.Normal('logc2', mu=pm.math.log(c_nfw(10**logMh)), sigma=0.11, testval=1.5)
    
    radius = data[0]
    vobs, e_vobs = data[1], data[2]
    vgas, vdisc, vbulge = data[3], data[4], data[5]
    
    vbar = np.sqrt(vgas**2 + 10**logml * vdisc**2 + 1.4 * 10**logml * vbulge**2)
    vmodel = np.sqrt(vc_dz(radius, 10**logMh, 10**logs1, 10**logc2)**2 + vbar**2)
    
    v_mod = pm.Normal('v_mod', mu=vmodel, tau=1./e_vobs**2, observed=vobs)
    

with pm.Model() as nfw_model:
    data  = pm.Data('data', rotcurve)
    logMh = pm.Uniform('logMh', lower=0., upper=20., testval=12.)
    logml = pm.Uniform('logml', lower=-2., upper=2., testval=-0.3)
    logc  = pm.Normal('logc', mu=pm.math.log(c_nfw(10**logMh)), sigma=0.11, testval=1.0)
    
    radius = data[0]
    vobs, e_vobs = data[1], data[2]
    vgas, vdisc, vbulge = data[3], data[4], data[5]
    
    vbar = np.sqrt(vgas**2 + 10**logml * vdisc**2 + 1.4 * 10**logml * vbulge**2)
    vmodel = np.sqrt(vc_nfw(radius, 10**logMh, 10**logc)**2 + vbar**2)
    
    v_mod = pm.Normal('v_mod', mu=vmodel, tau=1./e_vobs**2, observed=vobs)
```

```python
nsamples, ntune = 4000, 2000

print('Running MCMC on the DZ model')
with dz_model:
    dz_trace = pm.sample(draws=nsamples, step=pm.NUTS(), tune=ntune, discard_tuned_samples=True, return_inferencedata=False)

# print('Running MCMC on the NFW model')
# with nfw_model:
#     nfw_trace = pm.sample(draws=nsamples, step=pm.NUTS(), tune=ntune, discard_tuned_samples=True, return_inferencedata=False)
```

```python
dz_samples = pm.trace_to_dataframe(dz_trace)
dz_samples['logml'] = 10**dz_samples['logml']
dz_samples['logs1'] = 10**dz_samples['logs1']
dz_samples['logc2'] = 10**dz_samples['logc2']
dz_samples.rename(columns={'logml':'ml', 'logs1': 's1', 'logc2': 'c2'}, inplace=True)
dz_samples = dz_samples[['logMh', 's1', 'ml', 'c2']]
corner.corner(dz_samples, divergences=False, labels=[r"$\rm\log\,M_{halo}$", r"$\rm\ s_1$", r"$\rm M/L_\star$", r"$\rm c_2$"],
              quantiles=[.16,.50,.84], show_titles=True, smooth=1., smooth1d=1., label_kwargs={'fontsize':14})
# plt.savefig('output/samples.jpg')
plt.show()
```

```python
nfw_samples = pm.trace_to_dataframe(nfw_trace)
nfw_samples['logml'] = 10**nfw_samples['logml']
nfw_samples['logc'] = 10**nfw_samples['logc']
nfw_samples.rename(columns={'logml':'ml', 'logc': 'c'}, inplace=True)
nfw_samples = nfw_samples[['logMh', 'c', 'ml']]
corner.corner(nfw_samples, divergences=False, labels=[r"$\rm\log\,M_{halo}$", r"$\rm c$", r"$\rm M/L_\star$"],
              quantiles=[.16,.50,.84], title='NFW', show_titles=True, smooth=1., smooth1d=1., label_kwargs={'fontsize':14})
plt.show()
```

```python
mean_logMh, mean_s1, mean_ml, mean_c2 = np.mean(dz_samples['logMh']), np.mean(dz_samples['s1']), np.mean(dz_samples['ml']), np.mean(dz_samples['c2'])

radii, vrot_obs, evrot_obs, vgas_obs, vdisc_obs, vbulge_obs, _, _ = rotcurve
vstar_obs = np.sqrt(mean_ml * vdisc_obs**2 + 1.4 * mean_ml * vbulge_obs**2)
vdm = vc_dz(radii, 10**mean_logMh, mean_s1, mean_c2)
vtot = np.sqrt(vstar_obs**2 + vgas_obs**2 + vdm**2)

vms = [np.sqrt(dz_samples['ml'][k]*vdisc_obs**2 + 1.4*dz_samples['ml'][k]*vbulge_obs**2 + vgas_obs**2 +
                        vc_dz(radii, 10**dz_samples['logMh'][k], dz_samples['s1'][k], dz_samples['c2'][k])**2) for k in range(len(dz_samples))]

fig, ax = plt.subplots()

ax.errorbar(radii, vrot_obs, yerr=evrot_obs, fmt='.', mfc='none', label=r"$\rm V_{rot}$", c='k')
ax.plot(radii, vgas_obs, color='royalblue', ls=":", label=r"$\rm V_{gas}$")
ax.plot(radii, vstar_obs, color='goldenrod', dashes=[8, 4], label=r"$\rm V_\star$")
ax.plot(radii, vdm, c='darkmagenta', ls="-.", label=r"$\rm V_{DM}$")
ax.plot(radii, vtot, c='crimson', label=r"$\rm model$")
ax.fill_between(radii, np.median(vms,axis=0)+np.std(vms,axis=0), np.median(vms,axis=0)-np.std(vms,axis=0),
                color='crimson', alpha=0.3, label=r'$\rm error$')
ax.set_xlabel(r"$\rm Radius\,(kpc)$", fontsize=16);
ax.set_ylabel(r"$\rm Velocity\,(km\,s^{-1})$", fontsize=16);
ax.legend(bbox_to_anchor=(-0.15, 1.06, 1.3, .102), fontsize=12, mode="expand", ncol=6)
```
