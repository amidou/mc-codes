## Utility functions

```python
# constants
G, H = 4.301e-9, 67.6
Dc = 200.
b, g = 2., 3. 
def c_func(M): # from Dutton & Maccio' (2014) -- with c_200
    return 10.**(0.905 - 0.101 * (np.log10(M*H/100.)-12))

def c2_func(M):
    return 11.5 * (1. + np.power(M/(1.05e12), 0.65))

def fc(x):
    return np.log(1+x)-x/(1+x)

def Vvir(Mh):
    return np.sqrt((Dc*(H)**2/2)**(1./3.) * (G*Mh)**(2./3.))

def Mvir(Vh):
    return np.sqrt(2./Dc)/G/H * Vh**3

def Rvir(Mh):
    rho_c = 3. * (H)**2 / (8. * np.pi * G)
    rho_hat = 4. / 3. * np.pi * Dc * rho_c
    return 1e3 * np.power(Mh / rho_hat, 1./3.)

def params_from_s1c2(s1,c2,r1,rvir,mvir):
    x12=np.sqrt(r1/rvir)
    c12=np.sqrt(c2)
    a=(1.5*s1-2.*(3.5-s1)*x12*c12)/(1.5-(3.5-s1)*x12*c12)
    c=((s1-2.)/((3.5-s1)*x12-1.5/c12))**2.
    return (c, a, b, g, rvir, mvir)

def mu(c, a, b, g):
    return np.power(1.+np.power(c,1./b),b*(g-a))/np.power(c,3.-a)
```

## Likelihood, prior and posterior

```python
# circular velocity of NFW halo
#def vc(R, Mh, cc):
#    return np.sqrt(Vvir(Mh)**2*Rvir(Mh)/R*fc(cc*R/Rvir(Mh))/fc(cc))

# circular velocity of Dekel Zaho halo
def vc(r, Mh, s1, c2):
    rvir = Rvir(Mh)
    r1 = 0.01 * rvir
    cc, a, b, g, rvir, Mh = params_from_s1c2(s1,c2,r1,rvir,Mh)
    rs = float(rvir)/cc
    x = r/rs
    Vvir2 = G*Mh/rvir
    return np.sqrt(Vvir2*mu(cc, a, b, g)*cc*np.power(x,2.-a)/np.power(1.+np.power(x,1./b),b*(g-a)))


# log-likelihood
def lnlike(theta, vobs, e_vobs, vgas, vdisc, vbulge, rad):
    # parameters: logMh, s1, logc2, logM/L
    logMh, s1, logc2, lml = theta
    vbar = np.sqrt(vgas**2 + (10**lml) * vdisc**2 + 1.4*(10**lml) * vbulge**2)
    v_mod = np.sqrt(vc(rad, 10**logMh, s1, 10**logc2)**2 + vbar**2)
    chisq = np.sum(((v_mod-vobs)**2/e_vobs**2))
    
    return -chisq/(len(vobs)-3)

# log-prior: flat on Mh and M/L, Gaussian on c
def lnprior(theta):  # gaussian prior on the c2(Mvir) relation from Freundlich et al. 2020 (Equation 46 & Table 1)
    logMh, s1, logc2, lml = theta
    if 0 < logMh < 20 and -2 < lml < 2 and -2 < s1 < 5:
#         c_mean, c_sigma = c_func(10**logMh), 0.11
        c2_mean, c2_sigma = c2_func(10**logMh), 0.6
        return -0.5*(logc2-np.log10(c2_mean))**2/c2_sigma**2-np.log(np.sqrt(2*np.pi)*c2_sigma)
    return -np.inf  

# log-probability
def lnprob(theta, vobs, e_vobs, vgas, vdisc, vbulge, rad):
    lp = lnprior(theta)
    lk = lnlike(theta, vobs, e_vobs, vgas, vdisc, vbulge, rad)
    if not np.isfinite(lp) or not np.isfinite(lk):
        return -np.inf
    return lp + lk
```

## Running the MCMC

```python
d = np.genfromtxt("NGC5055_rotmod.dat")

rad = d[:, 0]
vobs, e_vobs = d[:, 1], d[:, 2]
vgas, vdisc, vbulge = d[:, 3], d[:, 4], d[:, 5]

dirin="./output/"

# MCMC analysis
# using 20 chains (walkers) and evaluating 4000 steps
ndim, nwalkers, nstep = 4, 100, 3000

# initial guess position
pos = [(12,1.0,0.5,-0.1) + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]

# run the MCMC sampler
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=(vobs, e_vobs, vgas, vdisc, vbulge, rad))
sampler.run_mcmc(pos, nstep)

# cut initial steps (burn-in phase)
burnin = 500
samples = sampler.chain[:, burnin:, :].reshape((-1, ndim))
samples[:,3] = 10**samples[:,3]  # transform to M/L

# get max-probability model
lnprobs = sampler.lnprobability[:, burnin:].reshape(-1)
max_prob = np.copy(samples[np.argmax(lnprobs),:])

# ###### Visualising the walkers ################
# samples_ = sampler.chain
# samples_ = samples_[:, 500:, :]

# samples_ = samples_.T
# samples_[3] = 10**samples_[3]

# fig, axs = plt.subplots(4,1, figsize=(10,8), sharex=True)
# ylabels = [r"$\rm\log\,M_{halo}$", r"$\rm\ s_1$", r"$\rm\log\,c_2$", r"$\Upsilon_\star$"]
# for i in range(samples_.shape[0]):
#     axs[i].plot(samples_[i], ls='-', c='gray')
#     axs[i].set_ylabel(ylabels[i])
# plt.xlabel('n_steps')
# plt.savefig('walkers.jpg')
##############################################
```

```python
corner.corner(samples, labels=[r"$\rm\log\,M_{halo}$", r"$\rm\ s_1$", r"$\rm\log\,c_2$", r"$\Upsilon_\star$"],
              quantiles=[.16,.50,.84], show_titles=True, smooth=1., smooth1d=1., label_kwargs={'fontsize':14});
# print(np.mean(samples[:,2]))
plt.savefig(dirin+"NGC5055_sample_DZ.png")
```

```python
fig, ax = plt.subplots()

# Vstar & VDM of the best model
vstar = np.sqrt(max_prob[3] * vdisc**2 + 1.4*max_prob[3] * vbulge**2)
vdm = vc(rad, 10**max_prob[0], max_prob[1], 10**max_prob[2])
# array of curves for each sampled model
vms = np.array([np.sqrt(samples[k,3]*vdisc**2 + 1.4*samples[k,3]*vbulge**2 +
                        vgas**2 + 
                        vc(rad, 10**samples[k,0], samples[k,1], 10**samples[k,2])**2) 
                for k in range(len(samples))])

ax.errorbar(rad, vobs, yerr=e_vobs, fmt='.', mfc='none', label=r"$\rm V_{rot}$", c='k')
ax.plot(rad, vgas, color='royalblue', ls=":", label=r"$\rm V_{gas}$")
ax.plot(rad, vstar, color='goldenrod', dashes=[8, 4], label=r"$\rm V_\star$")
ax.plot(rad, vdm, c='darkmagenta', ls="-.", label=r"$\rm V_{DM}$")
ax.plot(rad, np.sqrt(vstar**2 + vgas**2 + vdm**2), c='crimson', label=r"$\rm model$")
ax.fill_between(rad, np.median(vms,axis=0)+np.std(vms,axis=0), np.median(vms,axis=0)-np.std(vms,axis=0),
                color='crimson', alpha=0.3)
ax.set_xlabel(r"$\rm Radius/kpc$", fontsize=16);
ax.set_ylabel(r"$\rm Velocity/km\,s^{-1}$", fontsize=16);
ax.legend(bbox_to_anchor=(-0.1, 1.06, 1.2, .102), fontsize=13, mode="expand", ncol=5)
plt.savefig(dirin+"NGC5055_massmodel_DZ.png")
```
